```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_study = pd.merge(study_results, mouse_metadata, how="left", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview
mouse_study.head()
```

```python
# Checking the number of mice.
len(mouse_study["Mouse ID"].unique())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicate_mice = mouse_study[mouse_study.duplicated(["Mouse ID", "Timepoint"], keep=False)]
duplicate_mouse_id = duplicate_mice["Mouse ID"].unique()
duplicate_mouse_id
```

```python
# Optional: Get all the data for the duplicate mouse ID.
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
mouse_study_clean = mouse_study[~ mouse_study["Mouse ID"].isin(duplicate_mouse_id)]
mouse_study_clean.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(mouse_study_clean["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
drug_group = mouse_study_clean.groupby("Drug Regimen")
mean = drug_group["Tumor Volume (mm3)"].mean()
median = drug_group["Tumor Volume (mm3)"].median()
var = drug_group["Tumor Volume (mm3)"].var()
std = drug_group["Tumor Volume (mm3)"].std()
sem = drug_group["Tumor Volume (mm3)"].sem()

drug_tumor_df = pd.DataFrame({"Mean": mean, "Median": median, "Variance": var, "Standard Deviation": std, "SEM": sem})
drug_tumor_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_tp = mouse_study_clean.groupby("Mouse ID")["Timepoint"].max()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
final_tumor = pd.merge(max_tp, mouse_study_clean, how="left", on=["Mouse ID", "Timepoint"])
final_tumor.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drug_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_volumes = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

# Locate the rows which contain mice on each drug and get the tumor volumes
for drug in drug_list:
    subset = final_tumor[final_tumor["Drug Regimen"] == drug]
    
    # add subset 
    tumor_volumes.append(subset["Tumor Volume (mm3)"])

# Determine outliers using upper and lower bounds
for idx, volumes in enumerate(tumor_volumes):
    name = drug_list[idx]
    
    q1 = volumes.quantile(0.25)
    q3 = volumes.quantile(0.75)
    iqr = q3 - q1
    median = volumes.median()
    
    print(f"The lower quartile of {name} is: {round(q1, 2)}.")
    print(f"The upper quartile of {name} is: {round(q3, 2)}.")
    print(f"The interquartile range of {name} is: {round(iqr, 2)}.")
    print(f"The the median of {name} is: {round(median, 2)}.")
    
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    
    print(f"Values below {round(lower, 2)} could be outliers for {name}.")
    print(f"Values above {round(upper, 2)} could be outliers for {name}.")
    
    potential_outliers = final_tumor[
        (final_tumor["Drug Regimen"] == name) & 
        ((final_tumor["Tumor Volume (mm3)"] < lower) | 
         (final_tumor["Tumor Volume (mm3)"] > upper))
    ]
    
    num_outliers = potential_outliers.shape[0]
    print(f"The number of outliers for {name} is: {num_outliers}.")
    
    if num_outliers > 0:
        outlier_ids = potential_outliers["Mouse ID"].values
        print(f"The outliers are: {outlier_ids}.")
    print("")
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
cap_data = mouse_study_clean[mouse_study_clean["Drug Regimen"] == "Capomulin"][["Mouse ID", "Tumor Volume (mm3)", "Weight (g)"]]
avg_volume = cap_data.groupby("Mouse ID")["Tumor Volume (mm3)"].mean()
avg_weight = cap_data.groupby("Mouse ID")["Weight (g)"].mean()

plt.scatter(avg_weight, avg_volume)
plt.title("Capomulin: Mouse Weight vs. Avg Tumor Volume")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
slope, intercept, r_val, p_val, std_err = st.linregress(avg_weight, avg_volume)
regression_line = slope * avg_weight + intercept
equation = f"y = {round(slope, 2)}x + {round(intercept, 2)}"

plt.scatter(avg_weight, avg_volume)
plt.plot(avg_weight, regression_line, "r-")
plt.title("Capomulin: Mouse Weight vs. Avg Tumor Volume")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.annotate(equation, (22, 36), fontsize=10, color="red")

print(f"The correlation coefficient between mouse weight and average tumor volume is {round(r_val**2, 2)}.")

```
